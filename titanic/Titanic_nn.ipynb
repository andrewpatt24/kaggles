```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

alld = pd.concat([
        train[test.columns.tolist()+['Survived']],
        test
    ])
alld
```

```python
import re

alld['Ticket'] = alld['Ticket'].apply(lambda x: re.sub('[ 0-9.]+','',x))
```

```python
alld['Cabin'] = alld['Cabin'].apply(lambda x: x[0] if pd.notnull(x) else x)
alld.head()
```

```python
raw_vars = ['PassengerId','Survived','Age','Fare']

endd = alld[raw_vars]

endd['Age'] = endd['Age'].fillna(0)
```

## check age and fare

```python
import seaborn as sns

sns.distplot(endd['Age'])
```

```python
from sklearn.preprocessing import MinMaxScaler

endd['Age'] = MinMaxScaler().fit_transform(endd['Age'].fillna(0))
endd['Fare'] = MinMaxScaler().fit_transform(endd['Fare'].fillna(0))
```

```python
vars_to_flag = ['Embarked','Parch','Sex','SibSp','Pclass','Cabin','Ticket']

for v in vars_to_flag:
    endd = pd.concat([
            endd,
            pd.get_dummies(alld[v],prefix=v,dummy_na=True,drop_first=True)
        ],1)

endd.head()
```

```python
endd[pd.isnull(endd['Survived'])].to_csv('data/test_data_ready.csv',index=False)

endd[~pd.isnull(endd['Survived'])].to_csv('data/train_data_ready.csv',index=False)
```

## Modelling

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
test = pd.read_csv('data/test_data_ready.csv')
train = pd.read_csv('data/train_data_ready.csv')
test.shape, train.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(train[train.columns[2:]],train[train.columns[1]].apply(int))

from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=2)
dtc.fit(X_train[['Age','Fare']],y_train)

X_train = pd.concat([X_train.reset_index(drop=True),pd.DataFrame(dtc.decision_path(X_train[['Age','Fare']]).todense(),
             columns=['age_fare_leaf_'+str(x) for x in range(7)])],1)

X_test = pd.concat([X_test.reset_index(drop=True),pd.DataFrame(dtc.decision_path(X_test[['Age','Fare']]).todense(),
             columns=['age_fare_leaf_'+str(x) for x in range(7)])],1)

X_train.drop(['Age','Fare','age_fare_leaf_0'],axis=1,inplace=True)
X_test.drop(['Age','Fare','age_fare_leaf_0'],axis=1,inplace=True)

X_train = X_train.apply(lambda x: x.apply(int))
X_test = X_test.apply(lambda x: x.apply(int))
```

```python
from tensorflow.contrib.learn import DNNClassifier
from tensorflow.contrib.layers import real_valued_column
```

```python
# Specify that all features have real-value data
feature_columns = [real_valued_column("", dimension=X_train.shape[1])]

# Build 3 layer DNN with 10, 20, 10 units respectively.
classifier = DNNClassifier(feature_columns=feature_columns,
                                            hidden_units=[150],
                                            n_classes=3,
                                            model_dir="/tmp/iris_model")

# Fit model.
classifier.fit(x=X_train,
               y=y_train,
               steps=2000)

# Evaluate accuracy.
accuracy_score = classifier.evaluate(x=X_test,
                                     y=y_test)["accuracy"]
print('Accuracy: {0:f}'.format(accuracy_score))
```
